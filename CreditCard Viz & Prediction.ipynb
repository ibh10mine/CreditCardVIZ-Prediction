```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
dtnew = dt
```

```python
dtnew['Attrition_Flag'] = le.fit_transform(dtnew['Attrition_Flag'])
dtnew['Gender'] = le.fit_transform(dtnew['Gender'])
dtnew['Education_Level'] = le.fit_transform(dtnew['Education_Level'])
dtnew['Income_Category'] = le.fit_transform(dtnew['Income_Category'])
```

```python
dtnew
```

```python
X = dtnew.drop(columns=['CLIENTNUM','Dependent_count','Marital_Status','Attrition_Flag','Card_Category'])
y=dtnew['Attrition_Flag']
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
ypred=xgb.predict(X_test)
xacc = accuracy_score(ypred,y_test)
```

```python
# Use Randomforest model to predict.
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
ypred = rfc.predict(X_test)
racc = accuracy_score(ypred,y_test)
```

```python
svc = SVC()
svc.fit(X_train,y_train)
ypred = svc.predict(X_test)
svcacc = accuracy_score(ypred,y_test)
```

```python
model_accuracy = [['randomforest',racc],['xgboost',xacc],['svc',svcacc]]
```

```python
pd.DataFrame(model_accuracy,columns=['Model','Accuracy_Score'])
```
